```python
def get_fft_values(y_values, T, N, f_ratio=1):
    f_values = np.linspace(0, 1/(2.0 * T), N//2)
    fft_values_ = np.fft.fft(y_values)
    num_of_frequencies = int(f_ratio * len(f_values))
    fft_values = np.abs(fft_values_)
    return f_values[:num_of_frequencies], fft_values[:num_of_frequencies]
    
```

# Wavelets and their essence

We use the Fourier transform to transform a signal from its time-domain to frequency domain. The peaks in the frequency spectrum indicate the most occuring frequencies in the signal. The **larger** and **sharper** the peak is, the more prevalent a frequency is in the signal. The location (frequency-value) and height(amplitude) of the peaks in the frequency spectrum can be used as input for classifiers like Gradient Boosting etc.

</br>

## Moving from Fourier transform to Wavelet Transform

We have seen that by multiplying a signal with a series of sine-waves with different frequencies, we are able to determine the frequencies that are present in a signal.

The higher the amplitude of frequency, the higher the similarity between the signal and the sine-wave.

<div style="background: #808080; padding: 1.5rem 2rem; border-radius: 5px; color: #fff;">The problem with the Fourier transform is that is has good frequency resolution but poor time resolution. This essentially means that it is easy to detect the frequencies that occur within a signal however, its hard to find out where these frequencies occur.</div>

```python
# Let's check out an example
t_n = 1 # duration of the signal
N = 100000 # time steps
T = t_n / N # Period
f_s = 1 / T # Sampling rate


xa = np.linspace(0, t_n, num=N)
xb = np.linspace(0, t_n/4, (N//4))

frequencies = [4, 30, 60, 90]

y1a, y1b = np.sin(2*np.pi*frequencies[0]*xa), np.sin(2*np.pi*frequencies[0]*xb)
y2a, y2b = np.sin(2*np.pi*frequencies[1]*xa), np.sin(2*np.pi*frequencies[1]*xb)
y3a, y3b = np.sin(2*np.pi*frequencies[2]*xa), np.sin(2*np.pi*frequencies[2]*xb)
y4a, y4b = np.sin(2*np.pi*frequencies[3]*xa), np.sin(2*np.pi*frequencies[3]*xb)


# let's create 2 composite signals
signal_1 = y1a + y2a + y3a + y4a
signal_2 = np.concatenate([y1b, y2b, y3b, y4b])


# making the frequency spectrum
f_values1, fft_values1 = get_fft_values(signal_1, T, N, 0.003)
f_values2, fft_values2 = get_fft_values(signal_2, T, N, 0.003)


# plotting the frequency spectrum
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
ax[0,0].plot(xa, signal_1)
ax[0,1].plot(f_values1, fft_values1)
ax[1,0].plot(xa, signal_2)
ax[1,1].plot(f_values2, fft_values2)
```

<br/>

We can see that both frequency spectra look the same, so how do we distinguish between both signals? 

Well, we can use the **Short time Fourier transform**. It works by splitting the signals into several parts of equal length and computing the Fourier transform of the signal within that window however, it suffers from something called the Uncertainty principle.

The smaller the window size, the more sure we are about where a certain frequency occured but less about the frequency itself. The larger the window, the more we will know about the frequency value but less about time.

<b style="font-size: 1.5rem">So what next?</b>

## The Wavelet Transform

The wavelet transform has a high resolution in both time and frequency domains. It's accomplished by using wavelets of different scales. We first analyse the signal with a large scale to observe "large features" and subsequently use a small scale to observe smaller features.

<img src="resolution.png" />

**SUMMARY**

**Time series** has a high time resolution but zero frequency resolution. **Fourier transform** has a high frequency resolution and zero time resolution. **Short time fourier transform** has a medium size resolution in both domains.

**Wavelet transform**

- For small frequency values a high resolution in the frequency domain, low resolution in the time-domain.

- For large frequency values, a low resolution in the frequency domain, high resolution in the time domain.

<br/>

## How does the Wavelet transform work anyways?

The fourier transform uses a bunch of sines and cosines to identify the presence of frequencies whereas the **wavelet transform** uses a bunch of wavelets, each with a different scale.

<br/>

<div style="background: #808080; padding: 1.5rem 2rem; border-radius: 5px; color: #fff;">Wavelet means small wave</div>

<img src="wavelets.png" />

The 2D output of the wavelet transform is the time-scale representation of the signal in the form of a **scaleogram**

The wavelet is localised in time, thus we can convolute the wavelet with the signal at different points in time. We start at the beginning of the signal and slowly move the wavelet towards the end of the signal. The process is called **Convolution**.

The mother wavelet is used at first. Then its scaled subsequently to create daughter wavelets.

<br/>

But what are **SCALES** though?

Well, just like how Fourier transforms work with frequencies, the Wavelet transform works with **scales**. It is possible to convert scales to pseudo-frequencies with equation below.

$f_a = \frac {f_c}{a}$

Where:

$f_a$ = Pseudo-frequency

$f_c$ = Central frequency of the Mother wavelet

$a$ = scaling factor

<div style="background: #808080; padding: 1.5rem 2rem; border-radius: 5px; color: #fff;">
We can see that large values of a (long wavelets) correspond with smaller frequencies. So by scaling the wavelet in the time-domain, we analyze smaller frequencies properly (high frequency resolution). The converse is true for small values of a.</div>

<br/>


<br/>

## Wavelet Families

There are many different types of wavelet families because for each family, some trade-off has been made about how compact and smooth the wavelet looks like.

```python
import pywt
print(pywt.families(short=False))
```

Each type of wavelets has a different shape, smoothness and compactness and is useful for different applications. Since there are only 2 mathematical conditions a wavelet has to satisfy, it's pretty easy to create one.

*Mathematical Conditions*
1. A wavelet must have finite energy.
2. A wavelet must have a zero mean.

**Finite Energy** means that its localised in both time and frequency. It is integrable and inner product between the wavelet and the signal always exists.

```python
# plotting one of the wavelets
plt.figure(figsize=(15, 10))
wavelet = pywt.ContinuousWavelet("morl") # Continuous wavelet
wavelet_function, x_values = wavelet.wavefun()
plt.plot(x_values, wavelet_function)
plt.ylabel("Continuous Wavelet", fontsize=16)
plt.title("Moriet Wavelet")
plt.show()
```

# Continuous Wavelet Transform

As we have seen before, the wavelet transform comes in two different and distinct flavors namely:

1. The continuous wavelet transform
2. The discrete wavelet transform

<div style="font-size: 2rem; text-align:center">$X_w(a,b) = \frac{1}{|a|^{1/2}}\int^\infty_{-\infty}x(t)\bar{\psi}(\frac{t - b}{a}) dt$</div>

Where $\psi(t) = $ the continuous mother wavelet which gets scaled by a factor of $a$ and translated by a factor of $b$.

The values of the scaling and translation factors are continuous, which means there can be an **infinite** number of wavelets. You can scale the mother wavelet with a factor of 1.5, 1.51, 1.511, and so on.

<br/>

<br/>

# The Discrete Wavelet Transform

The main difference between the **discrete wavelet transform** and CWT is that DWT uses discrete values for the scaling and translation factors. 

The translation factors increases integer values whereas the scaling factor increases in powers of two.

In practice, the DWT is mostly implemented as a filter bank. This means that it is implemented as a cascade of high-pass and low-pass filters. This is done because filter bankls are very efficient in the splitting of a signal into several frequency sub-bands.

## The filtering process

1. To apply DWT to a signal, we start with the smallest scale. This means that we first analyze high frequency behavior.

2. In the second stage, the scale increases with a factor of two (which also means that the frequency reduces by a factor of two).

3. In the third stage, the scale increases by a factor of 4, and we are now analyzing frequency behaviour around a quarter of the maximum frequency. 

4. Now this goes on, until we have reached the maximum decomposition level.

### Decomposition Level

First, we need to understand that at each subsequent stage, the number of samples are also reduced by a factor of two. At lower frequency values, you will need less samples to satisfy the Nyquist rate so there is no need to keep the higher number of samples in the signal. Due to this downsampling, at some stage in the process, the number of samples in our signal will become smaller than the length of the wavelet filter and we will have reached maximum decomposition level.

<blockquote>The <b>Nyquist</b> frequency is the minimum rate at which a finite bandwidth signal needs to be sampled to retain all of the information.</blockquote>

For example, suppose we have a signal with frequencies up to 1000Hz. 

In the first stage, we split our signal into a low-frequency part and high-frequency part, i.e. 0-500Hz and 500-1000Hz.

In the second stage, we take the low frequency part again and split it into two parts (0-250) and (250-500).

This goes on until we have reached the level of refinement we need or until we run out of samples.

```python
# Let's visualize an example with the chirp signal
import pywt

x = np.linspace(0, 1, 2048)
chirp_signal = np.sin(250 * np.pi * x**2)

fig, ax = plt.subplots(figsize=(10, 2))
ax.set_title("Original chirp signal")
ax.plot(chirp_signal)
plt.show()

data = chirp_signal
waveletname = "sym5"


fig, axarr = plt.subplots(nrows=5, ncols=2, figsize=(12, 10))
for i in range(5):
    # the transform
    (data,  coeff_d) = pywt.dwt(data, waveletname)
    axarr[i,0].plot(data, 'r')
    axarr[i,1].plot(coeff_d, 'g')
    axarr[i,0].set_ylabel("Level {}".format(i+1), fontsize=14, rotation=90)
    axarr[i,0].set_yticklabels([])
    
    if i==0:
        axarr[i,0].set_title("Approximation coefficients", fontsize=14)
        axarr[i,1].set_title("Detail coefficients", fontsize=14)
    axarr[i,1].set_yticklabels([])
    
plt.tight_layout()
plt.show()
```

<br/>

As we can see, our original signal is now converted to several signals each corresponding to the different frequency bands. 

### The approximation coefficients

These are the output of the low pass filter. By applying DWT again on the approximation coefficients of the previous DWT, we get the wavelet transform of the next level.

### The detail coefficients

These are the outputs of the high pass filter

## Deconstructing a signal using DWT

In this section, we learn how to deconstruct a signal into its frequency sub-bands and reconstruct the signal subsequently.

There are two different ways to deconstruct a signal.

1. Apply <code>pywt.dwt()</code> on a signal to retrieve the approximation coefficients. Then apply DWT on the retrieve coefficients to get the second level coefficients and continue this process until you have reached a desired decomposition level.


2. We can also use <code>pywt.wavedec()</code> and directly retrieve all the of the detail coefficients up to some level $n$.

```python
(cA1, cD1) = pywt.dwt(signal_1, "db2", 'smooth')

# signal reconstruction
reconstructed_signal = pywt.idwt(cA1, cD1, 'db2', 'smooth')

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(signal_1, label="Signal")
ax.plot(reconstructed_signal, label="reconstructed signal", linestyle="--")
ax.legend(loc="upper left")
plt.show()
```

```python
# Let's try using pywt.wavedec()
coeffs = pywt.wavedec(signal_1, "db2", level=8)

# reconstructing the signal
reconstructed_signal = pywt.waverec(coeffs, "db2")

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(signal_1[:1000], label="signal")
ax.plot(reconstructed_signal[:1000], label="reconstructed signal", linestyle="--")
ax.legend(loc="upper left")
ax.set_title("deconstruction and reconstruction using wavedec")
plt.show()
```

# Removing noise (high frequency noise) using DWT

Remember that the detail coefficients represent the outputs of the high-pass filter. Therefore if you reconstruct the signal without these frequencies then you would have simply filtered out the frequency spectrum.

This can be done by using <code>pywt.threshold()</code>, which removes coefficient values higher than the given threshold.

```python
file_name = "acc_01210.csv"
df = pd.read_csv(file_name, header=None)
signal = df.loc[:, 4]
```

```python
?np.nanmax
```

```python
df.head()
```

```python
# creating a low pass filter
def lowpassfilter(signal, threshold=0.63, wavelet="db4"):
    thresh = np.nanmax(signal)*threshold
    coeff = pywt.wavedec(signal, wavelet, mode="per")
    
    # filtering high frequencies
    coeff[1:] = (pywt.threshold(i, value=thresh, mode="soft") for i in coeff[1:])
    
    # reconstructed signal (without high frequencies)
    reconstructed_signal = pywt.waverec(coeff, wavelet, mode="per")
    return reconstructed_signal
```

```python

fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(signal, color='b', alpha=0.5, label="original signal")
rec = lowpassfilter(signal, 0.4)
ax.plot(rec, 'k', label="DWT smoothing", linewidth=2)
ax.legend()
ax.set_title("Removing high frequency noise with DWT", fontsize=18)
ax.set_ylabel("Signal Amplitude", fontsize=16)
ax.set_xlabel("Sample No", fontsize=16)
plt.show()
```

We can see that by deconstructing the signal and setting some of the coefficients to zero, and reconstructing it agian, we can remove high frequency noise from a signal.

<div style="background: ###3377C6; padding: 1.5rem 2rem; border-radius: 5px; color: #fff; font-weight:bold">The advantage of DWT over other filtering methods is that there are different wavelets to choose from. Choosing a wavelet with a shape characteristic similar to some expected phenomena results in less of the phenomena being smoothed out.</div>

# Discrete Wavelet Classification

The idea is to use DWT to split the signal into the different frequency sub-bands, as many as needed or as many as possible. If the different types of signals exhibit different frequency idiosyncrasies, then this difference in behaviour has to be exhibited in one of the frequency sub-bands.

So we can generate some features about each of the sub-bands and use those features to train some classifier (random forest, gradient boosting, etc.). The classifier should then be able to distinguish between the different types of signals.

<img src="feature extraction.png" />

<br/>

## Generating features

There are several features that can be calculated for each sub-band however, few of the most frequently used are presented below.

1. Auto-regressive model coefficient values
2. (Shannon) Entropy values
3. Statistical features like:
 - variance
 - standard deviation
 - Mean
 - 25th percentile value
 - Mean crossing rate
 - The mean of the derivative 
 
 

```python
from collections import Counter
from scipy.stats import entropy

def calculate_entropy(list_values):
    counter_values = Counter(list_values).most_common()
    probabilities = [elem[1]/len(list_values) for elem in counter_values]
    shannon_entropy = entropy(probabilities)
    return shannon_entropy


def calculate_statistics(list_values):
    n5 = np.nanpercentile(list_values, 5)
    n25 = np.nanpercentile(list_values, 25)
    n75 = np.nanpercentile(list_values, 75)
    n95 = np.nanpercentile(list_values, 95)
    median = np.nanpercentile(list_values, 50)
    mean = np.nanmean(list_values)
    std = np.nanstd(list_values)
    var = np.nanvar(list_values)
    rms = np.nanmean(np.sqrt(list_values**2))
    
    return [n5, n25, n75, n95, median, mean, std, var, rms]


def calculate_crossings(list_values):
    zero_crossing_indices = np.nonzero(np.diff(np.array(list_values) > 0))[0]
    no_zero_crossings = len(zero_crossing_indices)
    mean_crossing_indices = np.nonzero(np.diff(np.array(list_values) > np.nanmean(list_values)))[0]
    no_mean_crossings = len(mean_crossing_indices)
    return [no_zero_crossings, no_mean_crossings]
    
    
def get_features(list_values):
    entropy = calculate_entropy(list_values)
    crossings = calculate_crossings(list_values)
    statistics = calculate_statistics(list_values)
    
    return [entropy] + crossings + statistics
```

The final <code>get_features</code> function generates 12 features for any list of values. So if one signal is decomposed into 10 different sub-bands, and we generate features for each sub-band, we will end up with $10*12 = 120$ features per signal.

```python
# Load the ECG data

def load_ecg_data(filename):
    raw_data = loadmat(filename)
    list_signals = raw_data['ECGData'][0][0][0]
    list_labels = list(map(lambda x: x[0][0], raw_data['ECGData'][0][0][1]))
    return list_signals, list_labels
    
```

```python
filename="ECGData/ECGData.mat"
data_ecg, labels_ecg = load_ecg_data(filename)
training_size = int(0.6*len(labels_ecg))
train_data_ecg = data_ecg[:training_size]
train_labels_ecg = labels_ecg[:training_size]
test_data_ecg = data_ecg[training_size:]
test_labels_ecg = labels_ecg[training_size:]
```

```python
np.unique(train_labels_ecg)
```

```python
# get ecg features
def get_ecg_features(ecg_data, ecg_labels, waveletname):
    list_features = []
    
    # labels are categorised into [0, 1, 2, 3... len(np.unique(ecg_labels))]
    list_unique_labels = list(set(ecg_labels))
    list_labels = [list_unique_labels.index(elem) for elem in ecg_labels]
    for signal in ecg_data:
        # remember that this returns wavelet coefficients
        list_coeff = pywt.wavedec(signal, waveletname)
        features = []
        for coeff in list_coeff:
            features += get_features(coeff)
            
        list_features.append(features)
        
    return list_features, list_labels
```

```python
X_train_ecg, Y_train_ecg = get_ecg_features(train_data_ecg, train_labels_ecg, "db4")
X_test_ecg, Y_test_ecg = get_ecg_features(test_data_ecg, test_labels_ecg, "db4")
```

```python
np.unique(Y_train_ecg)
```

```python
np.unique(Y_test_ecg)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

cls = GradientBoostingClassifier(n_estimators=2000)
cls.fit(X_train_ecg, Y_train_ecg)
train_score = cls.score(X_train_ecg, Y_train_ecg)
test_score = cls.score(X_test_ecg, Y_test_ecg)

print("Train score: {}".format(train_score))
print("Test score: {}".format(test_score))
```
